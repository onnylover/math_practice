## learning for prediction to survive
### select to be learned

#### data type conversion (Str -> Int/float)

```python
# Sex
raw_data
```

```python
# female=0, male=1
# same as lableendcoer/fit_transform
tmp =[]
for each in raw_data["Sex"]:
    if each == "female":
        tmp.append(0)
    elif each == "male":
        tmp.append(1)
    else:
        # for exception
        tmp.append(np.nan)
```

```python
raw_data["Sex"] = tmp
raw_data.head()
```

```python
raw_data.info()
```

```python
# data type int64 -> float64
raw_data['Survived'] = raw_data['Survived'].astype('float')
raw_data['Pclass'] = raw_data['Pclass'].astype('float')
raw_data['Sex'] = raw_data['Sex'].astype('float')
raw_data['SibSp'] = raw_data['SibSp'].astype('float')
raw_data['Parch'] = raw_data['Parch'].astype('float')
raw_data['Fare'] = raw_data['Fare'].astype('float')
raw_data.head()
```

```python
raw_data.info()
```

## build prediction model : Decision Tree

```python
raw_data.corr()
```

```python
train_pre = raw_data[["Pclass","Sex","Age","SibSp","Parch","Fare"]]
train_pre.head()
```

```python
train_pre.info()
```

```python
# filling missing value of x_train, x_test : fillna/imputer
meanAge = train_pre['Age'].mean()
train_pre['Age'] = train_pre['Age'].fillna(meanAge)
```

```python
raw_data.info()
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(train_pre,raw_data["Survived"],test_size=0.1, random_state=123, stratify=raw_data["Survived"])
```

```python
temp = y_train.astype("int")
temp
```

```python
sns.countplot(temp)
```

```python
x_train.shape, x_test.shape
```

```python
# reset index
x_train = x_train.reset_index(drop=True)
x_test = x_test.reset_index(drop=True)
y_train = y_train.reset_index(drop=True)
y_test = y_test.reset_index(drop=True)
```

```python
x_train.head()
```

```python
print(x_train.shape)
print(x_test.shape)
```

```python
x_train.info()
```

```python
x_test.info()
```

```python
from sklearn.tree import DecisionTreeClassifier

# assignmnet
model = DecisionTreeClassifier(max_depth=3, random_state=123)
```

```python
# learn
model.fit(x_train,y_train)
```

```python
# Standard of predict
# Accuracy(정확도) : 전체 대비 정확하게 예측한 수량
# Recall(재현율) : 예측 필요한 수량중에 맞춘 수량
# precision(정밀도) : 예측한 값중에 정확한 수량
# F1 : Recall, Precision 의 조화평균 

from sklearn.metrics import accuracy_score

# predict
prediction = model.predict(x_test)
print("Accuracy : {}%".format(accuracy_score(y_test, prediction)*100))
```

```python
(prediction == y_test).mean()*100
```
