## evaluation tool for linear regression
 1. MSE(Mean Squared Error) : 평균제곱오차 (차이가 두드러지도록)
 <br><br>
    ${(\frac{1}{n})\sum_{i=1}^{n}(y_{i} - x_{i})^{2}}$
    <br><br>
 2. MAE(Mean Absolute Error) : 평균절대오차 (실질적인 차이)
 <br><br>
   $(\frac{1}{n})\sum_{i=1}^{n}\left | y_{i} - x_{i} \right |$
  <br><br>
 3. RMSE(Root Mean Squared Error) : MSE에 루트처리
 <br><br>
    $\sqrt{(\frac{1}{n})\sum_{i=1}^{n}(y_{i} - x_{i})^{2}}$

```python
# example
pred = np.array([3,4,5])
target = np.array([2,3,4])
```

```python
def cal_MSE(pred,target):
    return ((pred-target)**2).mean()

print(cal_MSE(pred,target))
```

```python
def cal_MAE(pred,target):
    return np.abs(pred-target).mean()

print(cal_MAE(pred,target))
```

```python
def cal_RMSE(pred,target):
    return np.sqrt((pred-target)**2).mean()

print(cal_RMSE(pred,target))
```

```python
# evaluation tool from scikit-learn 
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
print(cal_MSE(pred,target))
print(mean_squared_error(pred,target))
```

```python
print(cal_MAE(pred,target))
print(mean_absolute_error(pred,target))
```

```python
print(cal_RMSE(pred,target))
print(mean_squared_error(pred,target,squared=False))
```

## Regularization
### prevent over/under-fitting
 1. L1 : Lasso (1차함수, 절대값)
 <br>
  depend size of alphad : large, small
 <br><br>
 $Error=MSE+α|w|$
 <br><br>
 2. L2 : Ridge (2차함수, 제곱)
 <br><br>
 $Error=MSE+αw^2$

#### L2

```python
from sklearn.linear_model import Ridge
```

```python
# large num is big regularization
alphas = [100,10,1,0.1,0.01,0.001]
```

```python
for alpha in alphas:
    # assignment
    ridge = Ridge(alpha=alpha)
    # learn
    ridge.fit(x_train,y_train)
    # predict
    pred = ridge.predict(x_test)
    mse_eval("Ridge(alpha)={}".format(alpha), pred, y_test)
```

```python
# same num as x_train columns
# linear regression 의 기울기 값 (합성곱을 위한 학습된 가중치의 갯수)
ridge.coef_
```

```python
x_train.columns
```

```python
def plot_coef(columns, coef):
    coef_df = pd.DataFrame(list(zip(columns, coef)))
    coef_df.columns=['feature', 'coef']
    coef_df = coef_df.sort_values('coef', ascending=False).reset_index(drop=True)
    
    fig, ax = plt.subplots(figsize=(9, 7))
    ax.barh(np.arange(len(coef_df)), coef_df['coef'])
    idx = np.arange(len(coef_df))
    ax.set_yticks(idx)
    ax.set_yticklabels(coef_df['feature'])
    fig.tight_layout()
    plt.show()
```

## Ensemble (앙상블)
### thru several ML model, find out the best answer
### do assignment few ML model, reference and mean result
 1. voting() : thru vote
 2. bagging : repeat make samples, and get result
   <br> ex) randomforest (based on DecisionTree)
 3. boosting : use previous result's gap, make it better and modify weight
 4. stacking : base on more than one predict model, predict again meta model (??) 여러모델을 기반으로 예측된 결과를 통해 meta 모델을 다시한번 예측


```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import ElasticNet
# poly ElasticNet
poly_pipeline = make_pipeline(
    PolynomialFeatures(degree=2, include_bias=False),
    StandardScaler(),
    ElasticNet(alpha=0.1, l1_ratio=0.2)
)
```

```python
poly_pred = poly_pipeline.fit(x_train, y_train).predict(x_test)
mse_eval("Poly ElasticNet", poly_pred, y_test)
```
