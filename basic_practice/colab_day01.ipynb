```python
from sklearn.datasets import load_digits
import matplotlib.pyplot as plt
import numpy as np
import matplotlib as mpl
mpl.rc('font', family='AppleGothic')

digits = load_digits()
sample = [0, 10, 20, 30, 1, 11, 21 ,31]
d = []
for i in range(8):
    d.append(digits.images[sample[i]])

plt.figure(figsize=(8,2))
for i in range(8):
    plt.subplot (1,8,i+1)
    plt.imshow(d[i], interpolation='nearest', cmap=plt.cm.bone_r)
    plt.grid(False)
    plt.xticks([])
    plt.yticks([])
    plt.title(f"image{i+1}")
plt.suptitle("Digits 0&1 이미지")
plt.tight_layout()
plt.show()

D = np.array(d)
print(D.shape)
 
```

```python
v =[]
for i in range(8):
    v.append(d[i].reshape(64,1))

V = np.array(v)
print(V.shape)

plt.figure(figsize=(8,3))
for i in range(8):
    plt.subplot (1,8,i+1)
    plt.imshow(v[i], aspect=0.4, interpolation='nearest', cmap=plt.cm.bone_r)
    plt.grid(False)
    plt.xticks([])
    plt.yticks([])
    plt.title(f"image{i+1}")
plt.suptitle("vector IMAGE")
plt.tight_layout(w_pad=7)
plt.show()

```

```python
from scipy import misc

img_rgb = misc.face()
img_rgb.shape

plt.subplot(221)
#plt.imshow(img_rgb, cmap=plt.cm.hsv)
plt.imshow(img_rgb, cmap=plt.cm.jet)
plt.axis("off")
plt.title("RGB color img")

plt.subplot(222)
plt.imshow(img_rgb[:,:,0], cmap=plt.cm.autumn)
plt.axis("off")
plt.title("RED color img")

plt.subplot(223)
plt.imshow(img_rgb[:,:,1], cmap=plt.cm.summer)
plt.axis("off")
plt.title("GREEN color img")

plt.subplot(224)
plt.imshow(img_rgb[:,:,2], cmap=plt.cm.winter)
plt.axis("off")
plt.title("BLUE color img")


plt.show()
```
