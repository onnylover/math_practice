```python
import glob

path = "/Volumes/Geozedo60/Python/bit/datamath/references/train/*.jpg"
file_list = glob.glob(path)

print("file_list : {}".format(file_list))
```

```python

topCount = 0
crossCount = 0
leftCount = 0
rightCount = 0

for file_name in file_list:
    # 카테고리에는 _ 앞에 값만 들어가고 나머지는 버림
    dirname, basename = os.path.split(file_name)
    category = basename.split("_")
    # print(category)
    
    if category[0] == "cross":
        crossCount += 1
    elif category[0] == "left":
        leftCount += 1
    elif category[0] == "right":
        rightCount += 1
    else:
        topCount += 1

print(f"crossCount: {crossCount}")
print(f"leftCount: {leftCount}")
print(f"rightCount: {rightCount}")
print(f"topCount: {topCount}")
```
